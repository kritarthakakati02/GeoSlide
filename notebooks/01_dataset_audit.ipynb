## 4. Missing Values

We quantify missing values across all columns, both as raw counts and as a percentage of total records. A visualization is included to make gaps in the data immediately apparent (or to confirm their absence).

```python
# Missing value counts and percentages per column
missing_count = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing Percentage (%)": missing_percent.round(2)
}).sort_values(by="Missing Count", ascending=False)

missing_df
```

```python
# Total missing values in the dataset
total_missing = df.isnull().sum().sum()
print(f"Total missing values in dataset: {total_missing}")
print(f"Overall missing percentage: {(total_missing / (df.shape[0] * df.shape[1])) * 100:.4f}%")
```

```python
# Visualize missing values per column
fig = px.bar(
    missing_df.reset_index().rename(columns={"index": "Column"}),
    x="Column",
    y="Missing Percentage (%)",
    title="Missing Values by Column (%)",
    text="Missing Count",
    color="Missing Percentage (%)",
    color_continuous_scale="Reds"
)
fig.update_layout(xaxis_tickangle=-60, template="plotly_white", height=500)
fig.show()
```

## 5. Duplicate Records

We check for fully duplicated rows in the dataset. Duplicate sensor readings could indicate logging errors or repeated transmissions and are relevant to note during the audit, even though no rows are removed at this stage.

```python
# Check for duplicate records
duplicate_count = df.duplicated().sum()
duplicate_percent = (duplicate_count / len(df)) * 100

print(f"Number of duplicate records: {duplicate_count}")
print(f"Percentage of duplicate records: {duplicate_percent:.2f}%")
```

## 6. Target Variable

The target column for the GeoSlide landslide prediction task is `Label`, where each record is classified into one of the landslide risk/event classes recorded by the wireless sensor network. We examine class counts and proportions to assess class balance.

```python
# Identify the target/label column
TARGET_COLUMN = "Label"

print(f"Target column: '{TARGET_COLUMN}'")
```

```python
# Class counts
class_counts = df[TARGET_COLUMN].value_counts().sort_index()
class_percent = (df[TARGET_COLUMN].value_counts(normalize=True).sort_index() * 100).round(2)

target_summary = pd.DataFrame({
    "Class": class_counts.index,
    "Count": class_counts.values,
    "Percentage (%)": class_percent.values
})

target_summary
```

```python
# Bar chart of target class distribution
fig = px.bar(
    target_summary,
    x="Class",
    y="Count",
    text="Count",
    color="Class",
    title="Target Variable (Label) Class Distribution",
    labels={"Class": "Label", "Count": "Number of Records"}
)
fig.update_traces(textposition="outside")
fig.update_layout(template="plotly_white", height=450, xaxis=dict(type="category"))
fig.show()
```

## 7. Numerical Features

This section provides descriptive statistics for all numerical features in the dataset (sensor readings such as rainfall, soil saturation, pore water pressure, seismic activity, etc.). These statistics describe the raw, unmodified distribution of each feature.

```python
# Identify numerical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Number of numerical columns: {len(numerical_cols)}")
print(numerical_cols)
```

```python
# Descriptive statistics for numerical features
numerical_stats = df[numerical_cols].describe().T
numerical_stats["range"] = numerical_stats["max"] - numerical_stats["min"]
numerical_stats.round(3)
```
